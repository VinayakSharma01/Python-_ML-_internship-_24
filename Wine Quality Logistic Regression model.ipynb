***MACHINE LEARNING***

```python
X=wq.drop(columns=['quality'])
Y=wq['quality']
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = TTS(X,Y,test_size=0.30,random_state=0)
```

```python
model = LOR(max_iter=1000)
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
DF_test_predict = X_test
DF_test_predict[" Y_test"] = Y_test
DF_test_predict[" Y_pred"] = Y_pred
DF_test_predict
```

```python
CM=confusion_matrix(Y_test,Y_pred)
AS = accuracy_score(Y_test,Y_pred)
RS = recall_score(Y_test,Y_pred)
PS = precision_score(Y_test,Y_pred)
F1 = f1_score(Y_test,Y_pred)
print(f"   accuracy_score = {AS}\nrecall_score={RS} \nprecision_score={PS}\nf1_score {F1}")
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(CM,annot=True,fmt='d',cmap='Reds',
            xticklabels=["Good","Bad"],
            yticklabels=["Good","Bad"]
            )
plt.xlabel("Predected")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.legend()

plt.show()
```
